```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ten_min_flow_speed_data = pd.read_csv("data/10sec_100_m_flow_speed_1hr.csv")
print(ten_min_flow_speed_data)
```

```python
flow = ten_min_flow_speed_data["Flow"]
speed = ten_min_flow_speed_data["Speed"]
ten_min_flow_speed_data["Density"] = flow/speed
print(ten_min_flow_speed_data["Space"].iloc[1000])
first_space_value = list(ten_min_flow_speed_data['Space'][1000:1003])
print(first_space_value)

filtered_data = ten_min_flow_speed_data#[ten_min_flow_speed_data["Space"].isin(first_space_value)]

print(filtered_data)
plt.figure(figsize=(8, 6))
plt.scatter(filtered_data['Density'], filtered_data['Flow'], c='blue', alpha=0.6)
plt.xlim(0, 6000)
plt.title('Density-Flow Diagram')
plt.xlabel('Density (vehicles per distance unit)')
plt.ylabel('Flow (vehicles per time unit)')
plt.grid(True)
plt.show()
```
