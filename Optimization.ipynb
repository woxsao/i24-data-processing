```python
from pyomo.environ import *

def joint_metanet_fit(v_hat, rho_hat, q_hat, T, l, initial_traffic_state, downstream_density, num_calibrated_segments):
    num_timesteps, num_segments = v_hat.shape
    
    model = ConcreteModel()
    model.t = RangeSet(0, num_timesteps-1)
    model.i = RangeSet(0, num_segments-1)
    model.segment_ix = RangeSet(0, num_calibrated_segments-1)

    # -------------------
    # Fundamental diagram parameters (decision variables)
    # -------------------
    model.rho_crit = Var(bounds=(1e-2, float(np.max(rho_hat))), initialize=np.median(rho_hat))
    model.v_free   = Var(bounds=(10, 150), initialize=60.0)
    model.a        = Var(bounds=(0.01, 10), initialize=1.0)

    # -------------------
    # METANET parameters (decision variables)
    # -------------------
    model.eta_high = Var(model.segment_ix, bounds=(10.0, 90.0), initialize=30.0)
    model.tau      = Var(model.segment_ix, bounds=(5/3600, 80.0), initialize=18/3600)
    model.K        = Var(model.segment_ix, bounds=(5.0, 50.0), initialize=40.0)

    # -------------------
    # State variables
    # -------------------
    model.v_pred   = Var(model.t, model.i, bounds=(0, 200),
                         initialize={(t,i): float(v_hat[t,i]) for t in range(num_timesteps) for i in range(num_segments)})
    model.rho_pred = Var(model.t, model.i, bounds=(1e-3, 300),
                         initialize={(t,i): float(rho_hat[t,i]) for t in range(num_timesteps) for i in range(num_segments)})
    model.q_pred   = Var(model.t, model.i, bounds=(1e-3, 10000),
                         initialize={(t,i): float(q_hat[t,i]) for t in range(num_timesteps) for i in range(num_segments)})

    # -------------------
    # Observed data as parameters
    # -------------------
    model.v_hat   = Param(model.t, model.i, initialize={(t,i): float(v_hat[t,i]) for t in range(num_timesteps) for i in range(num_segments)})
    model.rho_hat = Param(model.t, model.i, initialize={(t,i): float(rho_hat[t,i]) for t in range(num_timesteps) for i in range(num_segments)})
    model.q_hat   = Param(model.t, model.i, initialize={(t,i): float(q_hat[t,i]) for t in range(num_timesteps) for i in range(num_segments)})
    model.T       = Param(initialize=T)
    model.l       = Param(initialize=l)

    # -------------------
    # Equilibrium velocity function
    # -------------------
    def V_eq(m, rho):
        return m.v_free * exp(-1/m.a * (rho/m.rho_crit)**m.a)

    # -------------------
    # Initial conditions
    # -------------------
    model.init_constraints = ConstraintList()
    for i in range(num_segments):
        model.init_constraints.add(model.v_pred[0,i]   == v_hat[0,i].item())
        model.init_constraints.add(model.rho_pred[0,i] == rho_hat[0,i].item())

    # -------------------
    # Density dynamics
    # -------------------
    def rho_update(m, t, i):
        if t == 0: 
            return Constraint.Skip
        if i == 0:
            inflow = initial_traffic_state[t-1, 0]
            outflow = m.rho_pred[t-1,0] * m.v_pred[t-1,0]
            return m.rho_pred[t,0] == m.rho_pred[t-1,0] + m.T/m.l * (inflow - outflow)
        else:
            inflow = m.rho_pred[t-1,i-1] * m.v_pred[t-1,i-1]
            outflow = m.rho_pred[t-1,i] * m.v_pred[t-1,i]
            return m.rho_pred[t,i] == m.rho_pred[t-1,i] + m.T/m.l * (inflow - outflow)
    model.rho_dyn = Constraint(model.t, model.i, rule=rho_update)

    # -------------------
    # Velocity dynamics
    # -------------------
    def v_update(m, t, i):
        if t == 0: 
            return Constraint.Skip
        seg_group = i // (num_segments // num_calibrated_segments)
        tau = m.tau[seg_group]
        eta = m.eta_high[seg_group]
        K   = m.K[seg_group]
        rho = m.rho_pred[t-1,i]
        v_eq = V_eq(m, rho)

        if i == 0:
            prev_v = m.v_pred[t-1,0]
            next_rho = m.rho_pred[t-1,1] if num_segments > 1 else downstream_density[t-1]
        elif i == num_segments-1:
            prev_v = m.v_pred[t-1,i-1]
            next_rho = downstream_density[t-1]
        else:
            prev_v = m.v_pred[t-1,i-1]
            next_rho = m.rho_pred[t-1,i+1]

        current = m.v_pred[t-1,i]
        term1 = m.T/tau * (v_eq - current)
        term2 = m.T/m.l * current * (prev_v - current)
        term3 = (eta*m.T)/(tau*m.l) * (next_rho - rho) / (rho + K)
        return m.v_pred[t,i] == current + term1 + term2 - term3
    model.v_dyn = Constraint(model.t, model.i, rule=v_update)

    # -------------------
    # Flow definition
    # -------------------
    def q_update(m,t,i):
        return m.q_pred[t,i] == m.v_pred[t,i] * m.rho_pred[t,i]
    model.q_dyn = Constraint(model.t, model.i, rule=q_update)

    # -------------------
    # Objective: joint loss (v, rho, q)
    # -------------------
    def loss_fn(m):
        return sum(
            (m.v_pred[t,i] - m.v_hat[t,i])**2
            for t in m.t for i in m.i
        )
    model.loss = Objective(rule=loss_fn, sense=minimize)

    # -------------------
    # Solve
    # -------------------
    solver = SolverFactory("ipopt")
    solver.options['max_iter'] = 20000
    solver.solve(model, tee=False)

    return model
```

```python
def smooth_inflow(inflow, window_size=1):
    kernel = np.ones(window_size) / window_size
    smoothed = np.apply_along_axis(lambda m: np.convolve(m, kernel, mode='same'), axis=0, arr=inflow)
    return smoothed
    
l = DISTANCE
time_segments = np.arange(0, 3600, TIME_INT)
num_timesteps = len(time_segments)
T = TIME_INT / 3600
SPEED_LIMIT = 150  # KM/H

print(np.min(rho_hat))
v_hat = q_hat / rho_hat

num_calibrated_segments = v_hat.shape[-1] - 2
print("q_hat shape", num_calibrated_segments)

# Initial flow (upstream boundary condition)
initial_flow = q_hat[:, 0:1]
initial_flow = smooth_inflow(initial_flow)
initial_traffic_state = initial_flow

# Internal segment data
segment_rho_hat = rho_hat[:, 1:-1]       # shape (time, segment)
segment_v_hat   = v_hat[:, 1:-1]
segment_q_hat   = q_hat[:, 1:-1]

# Downstream boundary
downstream_density = rho_hat[:, -1:]
downstream_density = smooth_inflow(downstream_density)

np.save('initial_flow.npy', initial_flow)
np.save('downstream_density.npy', downstream_density)

# -------------------------------
# Call the JOINT optimization model
# -------------------------------
res_model = joint_metanet_fit(
    segment_v_hat,
    segment_rho_hat,
    segment_q_hat,
    T, l,
    initial_traffic_state,
    downstream_density,
    num_calibrated_segments
)

# -------------------------------
# Extract results
# -------------------------------
num_timesteps, num_segments = segment_v_hat.shape

v_pred_array   = np.zeros((num_timesteps, num_segments))
rho_pred_array = np.zeros((num_timesteps, num_segments))

for t in range(num_timesteps):
    for i in range(num_segments):
        v_pred_array[t, i]   = value(res_model.v_pred[t, i])
        rho_pred_array[t, i] = value(res_model.rho_pred[t, i])

# -------------------------------
# Print calibrated parameters
# -------------------------------
print("Optimized FD Parameters:")
print("rho_crit:", value(res_model.rho_crit))
print("v_free :", value(res_model.v_free))
print("a       :", value(res_model.a))
print("-----------")

for seg in range(num_calibrated_segments):
    print(f"Segment {seg}")
    print("tau     :", value(res_model.tau[seg]))
    print("K       :", value(res_model.K[seg]))
    print("eta_high:", value(res_model.eta_high[seg]))
    print("-----------")
```

```python
import numpy as np

def mape(flow_hat, flow_pred):
    """
    Compute the Mean Absolute Percentage Error (MAPE) between ground truth and prediction.

    Parameters:
        flow_hat (np.ndarray): Ground truth array of shape [t, i]
        flow_pred (np.ndarray): Predicted array of shape [t, i]

    Returns:
        float: The mean absolute percentage error (in percent)
    """
    # Avoid division by zero by masking out zero ground truth values
    mask = flow_hat != 0
    # print(flow_hat.shape)
    # print(flow_pred.shape)
    error = np.abs((flow_pred[mask] - flow_hat[mask]) / flow_hat[mask])
    return np.mean(error) * 100
def rmse(flow_hat, flow_pred):
    """
    Compute the Root Mean Squared Error (RMSE) between ground truth and prediction.

    Parameters:
        flow_hat (np.ndarray): Ground truth array of shape [t, i]
        flow_pred (np.ndarray): Predicted array of shape [t, i]

    Returns:
        float: The root mean squared error
    """
    error = flow_pred - flow_hat
    mse = np.mean(np.square(error))
    return np.sqrt(mse)
```
